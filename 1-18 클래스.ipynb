```python
# 클래스가 언제 사용하는지가 중요하다.

class Monster(): # 클래스는 중앙에서 관리한다라고 생각한다.
    hp = 100      # 속성 값을 제시
    alive = True

    def damage(self, attack):  # 속성 값을 건드리는 외부에서 부르는 작업
        self.hp = self.hp - attack   # self는 변수에 붙이는 거다.
        if self.hp <= 0:
            self.alive = False

    def status_check(self):  
        if self.alive == True:   # if self.alive: 으로 쓸수 있다. 
            print("살아있다!")
        else:
            print("죽었다!")


# 객체지향적인 것이다. 객체각각 가지고 있는거.
m1 = Monster()   # m1은 인스턴스이고 첫번째 몬스터
m1.damage(100)
m1.status_check()

m2 = Monster()   #m2는 인스턴스이고 두번째 몬스터이다.
m2.damage(90)
m2.status_check()
```
