```python
Map,fiter,lamda식

map - 리스트의 모든 원소를 조작하기
lamda - 불필요한 함수 줄이기. 반복적인 표기를 줄이고 요약하는것.
filter - 리스트의 모든 원소 중 특별한 것만 뽑기
```

```python
people = [
    {'name': 'bob', 'age': 19},
    {'name': 'carry', 'age': 38},
    {'name': 'john', 'age': 7},
    {'name': 'smith', 'age': 17},
    {'name': 'ben', 'age': 27},
    {'name': 'bobby', 'age': 57},
    {'name': 'red', 'age': 32},
    {'name': 'queen', 'age': 25}
]

def check_adult(person):
    if person["age"] <= 19:
        return "청소년"
    else:
        return "성인"

result = map(check_adult, people) # map은 people을 돌면서 check_adult(person)에다가 넣어라. 그래서 결과값을 리스트로 묶은 것이다. 여기서 주의할점은 리스트 함수를 써야 한다. map은 함수랑 짝꿍이네.
print(list(result))
```

```python
def check_adult(person):
    return "청소년" if person["age"] <= 19 else "성인"

result = map(check_adult, people)
list(result)
```

```python
# 한줄짜리 함수 만들기. 람다를 이용해서

result = map(lambda person: "성인" if person["age"]> 20 else "청소년", people)  # people을 돌면서 person에 적용해라 이다. 그래서 "성인" if person["age"]> 20 else "청소년"을 리턴해라.
list(result)
```

```python

result = filter(lambda x: x["age"] < 19 , people)  # people에 담겨진 데이터를 x에 넣고, x["age"] < 19만 뽑아라.

print(list(result))
```
