```python
# 문제 주민등록번호에서 남,여 구분하는 문제.
def check_gender(pin):
    pin = pin.split("-")[1][:1]
    if int(pin) % 2 == 0:
        print("여")
    else:
        print("남")

check_gender("15010101-1012345")
check_gender("15010101-2012345")
check_gender("15010101-4012345")


# 주의사항 ""이거는 다 문자열 처리된다.는 점 "1"
```

```python
# 문제 주민등록번호에서 남,여 구분하는 문제. 응용
def check_gender(pin):
    num = pin.split("-")[1][:1]
    print(num)
    if int(num) % 2 == 1:
        return ("남")
    else:
        return ("여")

check_gender("15010101-1012345")
check_gender("15010101-2012345")
check_gender("15010101-4012345")


(num) , sep=""  feat. 우현
(코드업 -> 백준)
```

```python
# return 값이 있는 함수
def return_name():
    return 'peter'

# return 값이 없는 함수
def print_name():
    print('peter')    # <<< 실제로는 여기까지 쓰여있습니다
    # return None     <<< 만약 함수에 return 이 정의되어 있지 않다면 파이썬 내부적으로 return None 이 함수 가장 마지막에 쓰여있다고 가정하게 됩니다.

print(return_name())
print_name()
```

```python
# retrun과 print의 차이 반환해주냐 안해주냐 차이다.
print("내 이름은 %s 입니다" % print_name())
print("내 이름은 %s 입니다" % return_name())
```
