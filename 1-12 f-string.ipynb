```python
# f-sting의 위력
scores = [
    {'name':'영수','score':70},
    {'name':'영희','score':65},
    {'name':'기찬','score':75},
    {'name':'희수','score':23},
    {'name':'서경','score':99},
    {'name':'미주','score':100},
    {'name':'병태','score':32}    
]

for s in scores:
    name = s["name"]
    score = s["score"]
    print(name+"의 점수는" +str(score)+"점입니다.")  # 주의 ""은 무조건 문자열이어야 한다. 그래서 str을 붙여 준것이다.
    print("==================")
    print(f"{name}은 점수가 {score}점입니다.") # f-sring은 상관없이 자료형에 상관없이 들어간다. f-string의 위력이다. 
```
